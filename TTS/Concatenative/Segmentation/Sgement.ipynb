```python
def process_textgrid(fname, speed=1):
    with open(fname, encoding='UTF-8') as f:
        f_lines = f.readlines()
        
        # yay spaghetti!
        xmins = map(lambda x: float(x.split('=')[1].strip()) * speed, filter(lambda x: x.startswith('            xmin'), f_lines))
        xmaxs = map(lambda x: float(x.split('=')[1].strip()) * speed, filter(lambda x: x.startswith('            xmax'), f_lines))
        texts = map(lambda x:       x.split('=')[1].strip()         , filter(lambda x: x.startswith('            text'), f_lines))
        
        
    return zip(xmins, xmaxs, texts)
        
    
def process_data(fname):
    with open(fname, encoding='UTF-8') as f:
        f_lines = f.readlines()
        
        seg_names = [line.split(' ')[1] for line in f_lines]
        ch_names =  [line[:-6] for line in seg_names]
        
    return zip(ch_names, seg_names)
```
